```python
import re
import string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import  WordCloud
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from sklearn.model_selection import train_test_split
import keras
import tensorflow as tf

pd.set_option("display.max_colwidth", 1000)
```

```python
df = pd.read_csv("D:/DATA BANGALORE/Data-NLP-Bahasa-Indonesia-main/review_hotel.csv", encoding='latin1')
df.head()
```

```python
df['category'].value_counts(dropna=False).sort_index().plot(kind='bar', rot='horizontal')
```

```python
text = " ".join(review for review in df.review_text)
wordcloud = WordCloud(background_color="white", max_words=10000, min_word_length=10).generate(text)
fig=plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
punctuations = re.sub(r"[!<_>#:)\.]", "", string.punctuation)

def punct2wspace(text):
    return re.sub(r"[{}]+".format(punctuations), " ", text)

def normalize_wspace(text):
    return re.sub(r"\s+", " ", text)

def casefolding(text):
    return text.lower()
```

```python
stemmer = StemmerFactory().create_stemmer()
```

```python
def preprocess_text(text):
    text = punct2wspace(text)
    text = normalize_wspace(text)
    text = casefolding(text)
    text = stemmer.stem(text)
    return text
```

```python
df["cleaned_text"] = df["review_text"].apply(preprocess_text)
df.head()
```

```python
target = df["category"]
features_text = df["cleaned_text"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_text, target, random_state=42)
```

```python
BATCH_SIZE = 32
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test))

train_dataset = train_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
for feat, tar in train_dataset.take(1):
    print(feat[:3])
    print(tar[:3])
```

```python
max_features = 10000     
embedding_dim = 16        
```

```python
encoder = keras.layers.TextVectorization(max_tokens=max_features)
encoder.adapt(train_dataset.map(lambda feat, tar: feat))
```

```python
encoder(feat)[:3]
```

```python
modelnlp = keras.Sequential()
modelnlp.add(encoder)
modelnlp.add(keras.layers.Embedding(
    input_dim=len(encoder.get_vocabulary()), 
    output_dim=embedding_dim, 
    mask_zero=True))

modelnlp.add(keras.layers.LSTM(100, return_sequences=True, dropout=0.5, recurrent_dropout=0.2))
modelnlp.add(keras.layers.LSTM(100))
modelnlp.add(keras.layers.Dense(8))
modelnlp.add(keras.layers.Dropout(0.2))
modelnlp.add(keras.layers.Dense(1, activation="sigmoid"))

modelnlp.compile(loss="binary_crossentropy", 
              optimizer="rmsprop",
              metrics=["acc"])
```

```python
modelnlp.summary()
```

```python
history = modelnlp.fit(train_dataset,
                    epochs=15, batch_size=BATCH_SIZE,
                    validation_data=test_dataset)
```

```python
scores = modelnlp.evaluate(test_dataset, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
print("loss: {}".format((scores[0])))
```

```python
plt.figure()
plt.title("Akurasi")
plt.plot(history.history["acc"])
plt.plot(history.history["val_acc"])

plt.figure()
plt.title("Loss")
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.show()
```

```python
contoh_review = ["lokasi jauh dari keramaian, kamar mandi sempit dan ac kurang dingin"]
prediksi = modelnlp.predict(contoh_review)
prediksi.squeeze() 
if prediksi.squeeze()>0.5:
    print(prediksi.squeeze())
    print("Review positif")
else:
    print(prediksi.squeeze())
    print("Review negatif")
```
